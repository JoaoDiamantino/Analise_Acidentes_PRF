```python
def insert_data_to_table(dataframe):
    
    # Substitua pelas suas informações de conexão
    connection = psycopg2.connect(database = 'desafio300',
                      host = 'localhost',
                      user = 'postgres',
                      password = '430084',
                      port = '5432'
                     )
    # Crie um cursor para executar comandos SQL
    cursor = connection.cursor()
    
    # Colunas do dataframe
    columns = dataframe.columns.tolist()
    
    # Crie a query de inserção
    insert_query = "INSERT INTO acidentes ({}) VALUES ({})".format(', '.join(columns), ', '.join(['%s'] * len(columns)))
    
    # Converta o dataframe para uma lista de tuplas
    data = [tuple(row) for row in dataframe.values]
    
    # Execute a inserção em loop
    try:
        for row in data:
            cursor.execute(insert_query, row)
        connection.commit()
        print("Dados inseridos com sucesso!")
    except Exception as e:
        connection.rollback()
        print("Erro ao inserir dados:", str(e))
    
    # Feche o cursor e a conexão
    cursor.close()
    connection.close()


# Chame a função para inserir os dados na tabela 'acidentes'
insert_data_to_table(data_final)
```

```python
def execute_query(query):
    
    conn = psycopg2.connect(database = 'desafio300',
                      host = 'localhost',
                      user = 'postgres',
                      password = '430084',
                      port = '5432'
                     )

    cur = con.cursor()

    cur.execute(query)
    columns = [desc[0] for desc in cur.description]
    rows = cur.fetchall()

    cur.close()
    conn.close()

    df = pd.DataFrame(rows, columns=columns)
    return df

# Consulta SQL
query = 'SELECT * FROM acidentes'

df = execute_query(query)
df.head()
```
